```python
import json
# Load raw OCR file
with open("/content/output/bill+lagoi_res.json", "r", encoding="utf-8") as f:
    raw = json.load(f)

texts = raw["rec_texts"]
scores = raw["rec_scores"]

# Gabungkan jadi list of dict
clean_output = [
    {"text": text, "confidence": round(score, 3)}
    for text, score in zip(texts, scores)
]

# Simpan ke file baru
with open("ocr_clean.json", "w", encoding="utf-8") as f:
    json.dump(clean_output, f, indent=2, ensure_ascii=False)

# Tampilkan contoh
for line in clean_output[:10]:
    print(f"{line['text']}  (conf: {line['confidence']})")
```

# Chunking

```python
import json

# 1. Load file JSON hasil struktur
with open("/content/parsed_structured.json", "r", encoding="utf-8") as f:
    structured_data = json.load(f)

# 2. Buat list of chunk dict
chunks = []
for item in structured_data:
    # Ambil nilai
    tgl = item.get("tanggal") or "null"
    nama = item.get("barang") or "unknown"
    qty = item.get("quantity")
    harga = item.get("harga")
    total = item.get("total")

    # Susun teks
    if qty is not None and harga is not None:
        teks = f"{nama} ({qty} x {harga} = {total}) pada {tgl}"
    else:
        teks = f"{nama} dengan total {total} pada {tgl}"

    chunks.append({"text": teks})

# 3. Simpan hasil ke file
with open("chunks_for_vector.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, indent=2, ensure_ascii=False)

# 4. Contoh output
for c in chunks:
    print(c["text"])
```

```python
import json

# Load file JSON hasil pemrosesan
with open("/content/parsed_structured.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Cek format data
for row in data:
    print(row)
```

```python
# Gabungkan ke bentuk kalimat agar bisa di-embed
texts = []
metadatas = []

for d in data:
    text = f"Pada tanggal {d['tanggal']}, terdapat pembelian {d['barang']} sebanyak {d['quantity']} dengan harga satuan {d['harga']} dan total {d['total']}."
    texts.append(text)
    metadatas.append(d)
```

```python
from langchain.vectorstores import FAISS
from langchain.embeddings import SentenceTransformerEmbeddings

# Gunakan model embedding ringan
embedding = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Simpan ke vectorstore
vectorstore = FAISS.from_texts(texts=texts, embedding=embedding, metadatas=metadatas)

# Simpan ke file
vectorstore.save_local("vector_ocr_db")
```

```python
# Load kembali vectorstore dari disk
db = FAISS.load_local("vector_ocr_db", embedding, allow_dangerous_deserialization=True)

# Lakukan semantic search
query = "Apa saja transaksi yang terjadi tanggal 2016-07-05?"
docs = db.similarity_search(query, k=3)

for d in docs:
    print(d.page_content)
```
